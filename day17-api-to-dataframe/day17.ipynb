```python
import pandas as pd
import requests
```

```python
url = "https://weatherbit-v1-mashape.p.rapidapi.com/forecast/3hourly"

querystring = {"lat":"35.5","lon":"-78.5","units":"imperial","lang":"en"}

headers = {
	"x-rapidapi-key": "660168453fmsh087750d26a82083p1099e7jsne0333be80e07",
	"x-rapidapi-host": "weatherbit-v1-mashape.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)
```

```python
temp_df1 = pd.DataFrame(response.json()['data'])[['uv','temp','ozone','wind_spd','pres','wind_dir']]
temp_df2 = pd.DataFrame(response.json())[['city_name','state_code','country_code']]
df = pd.concat([temp_df2, temp_df1], axis=1)
```
